```python
import pandas as pd
```

```python
url = "https://raw.githubusercontent.com/MauricioGomez-2520612022/datawarehouse-seguros/refs/heads/main/data/clientes.csv"
```

```python
df = pd.read_csv(url)
```

Segundo Bloque

```python
print("Primeras fila del Dataset")
display(df.head())
```

```python
print("Dimensiones del dataset (filas, columnas)")
print(df.shape)
```

```python
print("Valores nulos por columna")
print(df.isnull().sum())
```

```python
print("Registros duplicado")
print(df.duplicated().sum())
```

Tercer Bloque

```python
#Elimina espacios al inicio y al final en columnas de texto
for col in df.select_dtypes(include="object").columns:df[col]=df[col].str.strip()
```

```python
#Normalizar el nombre
df["nombre"]=df["nombre"].str.title()
```

```python
#Normalizar el email
df["email"]=df["email"].str.lower()
```

```python
#Normalizar genero
df["genero"]=df["genero"].str.upper()
```

```python
#Convertir fecha de nacimiento a formato fecha
df['fecha_nacimiento']=pd.to_datetime(df['fecha_nacimiento'],errors="coerce")
```

```python
#Normalizar ciudad
df["ciudad"]=df["ciudad"].str.title()
df['ciudad']=df['ciudad'].astype(str)
df['ciudad']=df['ciudad'].str.strip()
df['ciudad']=df['ciudad'].str.upper()
```

```python
#Limpieza de segmento
df['segmento']=df['segmento'].astype(str)
df['segmento']=df['segmento'].str.strip()
df['segmento']=df['segmento'].str.upper()
```
